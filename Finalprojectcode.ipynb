```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

data = np.genfromtxt('Finalprojectdata.csv',delimiter=",",skip_header=3,skip_footer=0)

obs=data[0:77,0]
time=data[0:77,1]
pressure=data[0:77,2]
temp=data[0:77,3]
salinity=data[0:77,4] 
nitratecon=data[0:77,5] 


fig, ax=plt.subplots()
ax.scatter(nitratecon, temp)
plt.grid()

plt.title('Nitrate Concentration vs. Temperature')
plt.xlabel('Nitrate Concentration [μmol/L]')
plt.ylabel('Temperature [Celsius]')

linregress(nitratecon, temp) 
```

```python
fig, ax=plt.subplots()
ax.scatter(nitratecon, pressure)
plt.grid()

plt.title('Nitrate Concentration vs. Pressure')
plt.xlabel('Nitrate Concentration [μmol/L]')
plt.ylabel('Pressure [dbar]')

linregress(nitratecon, pressure)
```

```python
fig, ax=plt.subplots()
ax.scatter(nitratecon, salinity)
plt.grid()
plt.title('Nitrate Concentration vs. Practical Salinity')
plt.xlabel('Nitrate Concentration [μmol/L]')
plt.ylabel('Practical Salinity')

linregress(nitratecon,salinity)
```
