```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

data = np.genfromtxt('finaldata.csv',delimiter=",",skip_header=1,skip_footer=0)

obs=data[0:39,0]
time=data[0:39,1]
AVGWVHT=data[0:39,2]
MAXWVHT=data[0:39,3]
MEANWVPD=data[0:39,4] 
RELWINDSPD=data[0:39,5] 


fig, ax=plt.subplots()
ax.scatter(RELWINDSPD, AVGWVHT)
plt.grid()

plt.title('Relative Wind Speed vs. Average Wave Height')
plt.xlabel('Relative Wind Speed [m/s]')
plt.ylabel('Average Wave Height [m]')

linregress(RELWINDSPD, AVGWVHT) 
```

```python
fig, ax=plt.subplots()
ax.scatter(RELWINDSPD, MEANWVPD)
plt.grid()

plt.title('Relative Wind Speed vs. Mean Wave Period')
plt.xlabel('Relative Wind Speed [m/s]')
plt.ylabel('Mean Wave Period [s]')

linregress(RELWINDSPD, MEANWVPD)
```

```python
fig, ax=plt.subplots()
ax.scatter(RELWINDSPD, MAXWVHT)
plt.grid()
plt.title('Relative Wind Speed vs. Maximum Wave Height')
plt.xlabel('Relative Wind Speed [m/s]')
plt.ylabel('Maximum Wave Height [m]')

linregress(RELWINDSPD,MAXWVHT)
```
